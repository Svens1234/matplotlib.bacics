```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.arange(0,10)
```

```python
y=2*x
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.title("Название графика");
```

```python
plt.plot(x,y)
plt.xlabel("Ось x")
plt.ylabel("Ось y")
plt.title("Название графика");
```

```python
plt.plot(x,y)
plt.xlabel("Ось x")
plt.ylabel("Ось y")
plt.xlim(0,6)
plt.ylim(0,15)
plt.title("Название графика");
```
